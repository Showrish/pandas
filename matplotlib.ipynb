```python
import matplotlib.pyplot as mp
import numpy as np
players=np.array(['Rohit','Kohli','Hardik','SriRam','Dhoni'])
runs=np.array([45,1,5,0,95])
oruns=np.array([98,56,12,0,43])
mp.plot(players,runs,'o',ls=':',mfc='r',mec='g',label='T20')
mp.plot(players,oruns,'o',mfc='b',ls='-',label='ODI')
mp.legend()
mp.xlabel('PLAYERS')
mp.ylabel('RUNS')
mp.show()
```

```python
import matplotlib.pyplot as mp
import numpy as nb
roll=nb.array([7,5,2,3,1])
gpa=nb.array([9.5,8.6,7.5,8.2,6.8])
mp.plot(roll,gpa,'o',ls=':',mfc='r')
mp.xlabel('RollNo')
mp.ylabel('GPA')
mp.show()
```

```python
import matplotlib.pyplot as mp
import numpy as np
n=np.array(['SaiRam','Mahesh','Dhanush','Rajiv','Rakesh'])
h=np.array([150,165,169,178,160])
w=np.array([95,100,72,70,90])
mp.plot(n,h,'o',ls=':',mfc='r')
mp.plot(n,w,'s',ls='-',mfc='b')
mp.show()
```
